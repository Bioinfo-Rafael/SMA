# 01 - Load each GSE to AnnData

One section per GSE. Run cell by cell, inspect, and collect the objects in the `adatas` dict. Raw counts go into `adata.X`; only mechanical obs fields (sample_id, source_file, ...) are set here. Biological metadata (genotype / treatment / condition) is set in **03**, not here.

GSE295514 (RDS) requires the R notebook to have produced the intermediate files first (see `notebooks/R/01_GSE295514_read_rds.ipynb`).

```python
import sys
from pathlib import Path

# locate the v2 project root (folder containing config/dataset_manifest.yaml)
ROOT = Path.cwd()
while not (ROOT / "config" / "dataset_manifest.yaml").exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
print("project root:", ROOT)

import manifest_utils as mf
man = mf.load_manifest()
paths = mf.project_paths(ROOT)
mf.ensure_dirs(paths)

import io_10x
import io_dense
import anndata_utils as au

adatas = {}
```

## GSE287569

10x .h5 per sample (TAR of H5)

```python
ds = mf.get_dataset(man, "GSE287569")
adata = io_10x.load_10x_h5_per_sample(ds, paths)
adatas["GSE287569"] = adata
print(adata.shape)
adata
```

```python
# quick look
display(adata.obs.head())
display(adata.var.head())
print('obs columns:', list(adata.obs.columns))
print('var columns:', list(adata.var.columns))
print('X:', type(adata.X), adata.X.dtype)
```
