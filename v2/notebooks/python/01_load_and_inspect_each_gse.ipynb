```python
from pathlib import Path
import gzip
from scipy.io import mmread

base = Path("/home/suzuki/Learn/SMA/v2/data/extracted/GSE167327")

mtx = base / "GSM5101954_counts.mtx.gz"
rows = base / "GSM5101954_counts.mtx.rownames.txt.gz"
cols = base / "GSM5101954_counts.mtx.colnames.txt.gz"

print("mtx exists:", mtx.exists(), mtx.stat().st_size / 1024 / 1024, "MB")
print("rows exists:", rows.exists(), rows.stat().st_size / 1024, "KB")
print("cols exists:", cols.exists(), cols.stat().st_size / 1024, "KB")

with gzip.open(rows, "rt") as f:
    rownames = [line.strip() for line in f if line.strip()]

with gzip.open(cols, "rt") as f:
    colnames = [line.strip() for line in f if line.strip()]

print("n rownames:", len(rownames))
print("n colnames:", len(colnames))
print("rownames head:", rownames[:5])
print("colnames head:", colnames[:5])

X = mmread(mtx).tocsr()
print("mtx shape:", X.shape)
print("nnz:", X.nnz)
print("sum:", X.sum())
```

```python
# ============================================================
# GSE167327
# MTX + rownames + colnames 形式を直接読む
# matrix: genes x cells
# AnnData: cells x genes
# ============================================================

from pathlib import Path
import gzip

import numpy as np
import pandas as pd
import scipy.sparse as sp
from scipy.io import mmread
import anndata as ad

ds = mf.get_dataset(man, "GSE167327")

# ----------------------------
# 1. file path
# ----------------------------
base = paths["extracted"] / "GSE167327"

mtx_path = base / "GSM5101954_counts.mtx.gz"
row_path = base / "GSM5101954_counts.mtx.rownames.txt.gz"
col_path = base / "GSM5101954_counts.mtx.colnames.txt.gz"

print("base:", base)
print("mtx:", mtx_path, mtx_path.exists())
print("rows:", row_path, row_path.exists())
print("cols:", col_path, col_path.exists())

if not mtx_path.exists():
    raise FileNotFoundError(mtx_path)
if not row_path.exists():
    raise FileNotFoundError(row_path)
if not col_path.exists():
    raise FileNotFoundError(col_path)


# ----------------------------
# 2. rownames / colnames
# ----------------------------
with gzip.open(row_path, "rt") as f:
    gene_names = [line.strip() for line in f if line.strip()]

with gzip.open(col_path, "rt") as f:
    cell_names = [line.strip() for line in f if line.strip()]

gene_names = [str(x) for x in gene_names]
cell_names = [str(x) for x in cell_names]

print("n genes from rownames:", len(gene_names))
print("n cells from colnames:", len(cell_names))
print("gene head:", gene_names[:5])
print("cell head:", cell_names[:5])


# ----------------------------
# 3. Matrix Marketを読む
# ----------------------------
X_gene_cell = mmread(mtx_path).tocsr()

print("raw mtx shape:", X_gene_cell.shape)
print("raw mtx nnz:", X_gene_cell.nnz)
print("raw mtx sum:", float(X_gene_cell.sum()))


# ----------------------------
# 4. shape validation & transpose
# ----------------------------
if X_gene_cell.shape == (len(gene_names), len(cell_names)):
    print("Detected orientation: genes x cells")
    X_cell_gene = X_gene_cell.T.tocsr()
elif X_gene_cell.shape == (len(cell_names), len(gene_names)):
    print("Detected orientation: cells x genes")
    X_cell_gene = X_gene_cell.tocsr()
else:
    raise ValueError(
        "Shape mismatch: "
        f"X={X_gene_cell.shape}, "
        f"genes={len(gene_names)}, cells={len(cell_names)}"
    )

print("AnnData X shape:", X_cell_gene.shape)


# ----------------------------
# 5. obsを作る
# ----------------------------
obs_index = [f"GSE167327_{c}" for c in cell_names]

obs = pd.DataFrame(index=pd.Index(obs_index, name=None))
obs["cell_id_original"] = cell_names
obs["cell_id"] = obs.index.astype(str)

obs["sample_id"] = "GSE167327"
obs["sample_label"] = "GSE167327_CD45_enriched"
obs["gsm_id"] = "GSM5101954"

obs["source_file"] = mtx_path.name
obs["source_accession"] = "GSE167327"
obs["parent_gse"] = "GSE167332"
obs["dataset_id"] = ds.get(
    "dataset_id",
    "GSE167327_sc_spinalcord_optn_cd45_enriched_indrop"
)

# 後段で共通化するための最低限metadata
obs["species"] = "Mus musculus"
obs["disease_area"] = "ALS_related_or_neuroinflammation"
obs["tissue"] = "spinal cord"
obs["region"] = "spinal cord"
obs["assay"] = "scRNA-seq"
obs["technology"] = "inDrop"
obs["enrichment"] = "CD45 enriched"
obs["disease_model"] = "Optn floxed/KO"
obs["data_status"] = "raw_or_filtered_count"
obs["processing_status"] = "GEO_supplementary_MTX"

# まだcondition/genotypeはGSM/論文側と照合して手動で入れる
obs["condition_original"] = "unknown"
obs["disease_status"] = "unknown"
obs["genotype"] = "unknown"
obs["treatment"] = "none_or_unknown"
obs["age"] = "unknown"
obs["age_month"] = "unknown"
obs["sex"] = "unknown"
obs["replicate"] = "unknown"


# ----------------------------
# 6. varを作る
# ----------------------------
var = pd.DataFrame(index=pd.Index(gene_names, name=None))
var["gene_symbol"] = var.index.astype(str)
var["gene_symbol_upper"] = var["gene_symbol"].str.upper()
var["gene_id"] = var["gene_symbol"]
var["ensembl_id"] = "unknown"
var["feature_type"] = "Gene Expression"


# ----------------------------
# 7. AnnData作成
# ----------------------------
adata = ad.AnnData(
    X=X_cell_gene.astype(np.float32),
    obs=obs,
    var=var,
)

adata.obs_names_make_unique()
adata.var_names_make_unique()

adata.uns["source_accession"] = "GSE167327"
adata.uns["parent_gse"] = "GSE167332"
adata.uns["dataset_id"] = obs["dataset_id"].iloc[0]
adata.uns["data_status"] = "raw_or_filtered_count"
adata.uns["note"] = (
    "GSE167327 CD45-enriched inDrop data. "
    "Loaded from GSM5101954_counts.mtx.gz with rownames/colnames files. "
    "Original MTX is genes x cells; transposed to cells x genes for AnnData."
)

adatas["GSE167327"] = adata


# ----------------------------
# 8. 確認
# ----------------------------
nb.summarize_adata(adata, name="GSE167327")
display(adata)

display(adata.obs.head())
display(adata.var.head())

_ = nb.show_obs_value_counts(adata, max_unique=30)

print("shape:", adata.shape)
print("X type:", type(adata.X))
print("X dtype:", adata.X.dtype)
print("X nnz:", adata.X.nnz if sp.issparse(adata.X) else np.count_nonzero(adata.X))
print("X sum:", float(adata.X.sum()))
```

## GSE206330

SoupX 処理済み + メタCSV（別扱い）

```python
# ============================================================
# GSE206330 修正版 loader: whitespace header + cell-id harmonization
# ============================================================

import gzip
import csv
from pathlib import Path

import numpy as np
import pandas as pd
import scipy.sparse as sp
import anndata as ad

ds = mf.get_dataset(man, "GSE206330")

raw_dir = paths["raw"] / "GSE206330"

count_path = raw_dir / "GSE206330_processed_SoupX_corrected.txt.gz"
meta_path = raw_dir / "GSE206330_processed_SoupX_corrected_metadata.csv.gz"

print("count_path:", count_path, count_path.exists())
print("meta_path :", meta_path, meta_path.exists())


# ----------------------------
# 1. metadataを読む
# ----------------------------
meta = pd.read_csv(meta_path, index_col=0)
meta.index = meta.index.astype(str)

print("metadata shape:", meta.shape)
print("metadata index head:", list(meta.index[:5]))
display(meta.head())


# ----------------------------
# 2. matrixのheaderだけ読む
# ----------------------------
with gzip.open(count_path, "rt") as f:
    header_line = f.readline().strip()

# whitespace区切りだが、cell IDがquoteされているので csv.reader を使う
cell_ids_matrix = next(csv.reader([header_line], delimiter=" ", quotechar='"'))

cell_ids_matrix = pd.Index(cell_ids_matrix.astype(str) if hasattr(cell_ids_matrix, "astype") else [str(x) for x in cell_ids_matrix])

print("n cell ids in matrix header:", len(cell_ids_matrix))
print("matrix cell ids head:", list(cell_ids_matrix[:5]))
print("metadata rows:", meta.shape[0])


# ----------------------------
# 3. matrix cell IDをmetadata IDに合わせる
# ----------------------------
# matrix: AAACCCAAGGGCTTCC.1M_CTRL
# meta  : AAACCCAAGGGCTTCC-1M_CTRL
#
# barcode中の . は基本ない想定で、最後の sample suffix separator を . -> - にする。
# より安全に「最初の .」だけ置換する。

cell_ids_harmonized = pd.Index([x.replace(".", "-", 1) for x in cell_ids_matrix])

print("harmonized cell ids head:", list(cell_ids_harmonized[:5]))

common = cell_ids_harmonized.intersection(meta.index)
print("common cells:", len(common))
print("common / matrix:", len(common), "/", len(cell_ids_harmonized))
print("common / meta  :", len(common), "/", meta.shape[0])

if len(common) == 0:
    raise ValueError("matrix header cell IDs and metadata index do not match after '.' -> '-' harmonization")

if len(common) != len(cell_ids_harmonized) or len(common) != meta.shape[0]:
    print("WARNING: not all cells match. Will subset to common cells.")
else:
    print("All cells matched.")


# ----------------------------
# 4. 大きなmatrixをchunkで読む
# ----------------------------
def load_gse206330_soupx_whitespace_header(
    count_path,
    meta,
    cell_ids_matrix,
    cell_ids_harmonized,
    dataset_id="GSE206330_sc_cortex_sod1_facs_glia_soupx",
    source_accession="GSE206330",
    chunksize=500,
):
    """
    GSE206330_processed_SoupX_corrected.txt.gz:
      line 1: "cell1" "cell2" ...
      line 2+: "gene" value value ...
    を読む。

    入力matrixは genes x cells。
    出力AnnDataは cells x genes。
    """

    # metadataと共通する列だけ残すためのmask
    common = cell_ids_harmonized.intersection(meta.index)
    keep_mask = cell_ids_harmonized.isin(common)

    kept_matrix_ids = cell_ids_matrix[keep_mask]
    kept_harmonized_ids = cell_ids_harmonized[keep_mask]

    print("kept cells:", len(kept_matrix_ids))

    # metadataはmatrix列順に並べる
    obs = meta.loc[kept_harmonized_ids].copy()
    obs.index = kept_harmonized_ids.astype(str)

    X_blocks = []
    gene_names = []

    # header=0で1行目を列名として読み、index_col=0でgene名をindexにする
    # ただしpandasがheader長とdata長の差をどう扱うかが微妙な場合があるので、
    # namesを明示して読む。
    col_names = ["gene_symbol"] + list(cell_ids_matrix)

    reader = pd.read_csv(
        count_path,
        sep=r"\s+",
        engine="python",
        header=0,
        names=col_names,
        index_col=0,
        chunksize=chunksize,
    )

    for i, chunk in enumerate(reader):
        # chunk: genes_chunk x cells
        if i == 0:
            print("first chunk raw shape:", chunk.shape)
            print("first chunk index head:", list(chunk.index[:5]))
            print("first chunk columns head:", list(chunk.columns[:5]))

        # common cellsだけ残す。列名はmatrix側のdot形式。
        chunk = chunk.loc[:, kept_matrix_ids]

        gene_names.extend(chunk.index.astype(str).tolist())

        # float値あり。SoupX correctedなので整数raw countではない。
        arr = chunk.to_numpy(dtype=np.float32, copy=False)

        # AnnDataは cells x genes
        X_blocks.append(sp.csr_matrix(arr.T))

        if (i + 1) % 10 == 0:
            print(f"  chunks={i+1}, genes={len(gene_names)}")

    X = sp.hstack(X_blocks, format="csr")

    # obs整形
    obs["cell_id_original"] = obs.index.astype(str)
    obs["sample_id"] = obs["Sample"].astype(str) if "Sample" in obs.columns else "unknown"
    obs["sample_label"] = obs["sample_id"].astype(str)
    obs["source_file"] = Path(count_path).name
    obs["source_accession"] = source_accession
    obs["dataset_id"] = dataset_id
    obs["data_status"] = "processed_SoupX_corrected"
    obs["processing_status"] = "SoupX_corrected"

    # manifest metadataも入れる
    obs["species"] = "Mus musculus"
    obs["disease_area"] = "ALS"
    obs["tissue"] = "cerebral cortex"
    obs["region"] = "motor cortex-related cortical glia"
    obs["assay"] = "scRNA-seq"
    obs["technology"] = "10x Genomics"
    obs["enrichment"] = "FACS enriched astrocytes/microglia/oligodendrocytes"
    obs["disease_model"] = "SOD1G93A"

    # conditionなど
    if "Condition" in obs.columns:
        obs["condition_original"] = obs["Condition"].astype(str)
        obs["disease_status"] = obs["Condition"].map({
            "CTRL": "control",
            "SOD1": "ALS_model",
            "ALS": "ALS_model",
        }).fillna(obs["Condition"].astype(str))
    else:
        obs["condition_original"] = "unknown"
        obs["disease_status"] = "unknown"

    if "Age" in obs.columns:
        obs["age"] = obs["Age"].astype(str)
        obs["age_month"] = obs["Age"].astype(str).str.extract(r"(\d+)")[0]
    else:
        obs["age"] = "unknown"
        obs["age_month"] = "unknown"

    if "sex" not in obs.columns:
        obs["sex"] = "unknown"

    # obs_namesを一意化
    obs.index = [f"{source_accession}_{x}" for x in obs.index.astype(str)]
    obs["cell_id"] = obs.index.astype(str)

    # var
    var = pd.DataFrame(index=pd.Index(gene_names, name=None).astype(str))
    var["gene_symbol"] = var.index
    var["gene_symbol_upper"] = var["gene_symbol"].str.upper()
    var["gene_id"] = var["gene_symbol"]
    var["ensembl_id"] = "unknown"
    var["feature_type"] = "Gene Expression"

    adata = ad.AnnData(X=X, obs=obs, var=var)
    adata.var_names_make_unique()

    adata.uns["source_accession"] = source_accession
    adata.uns["dataset_id"] = dataset_id
    adata.uns["data_status"] = "processed_SoupX_corrected"
    adata.uns["note"] = "GSE206330 SoupX-corrected processed matrix; not raw UMI counts."

    return adata


adata_fixed = load_gse206330_soupx_whitespace_header(
    count_path=count_path,
    meta=meta,
    cell_ids_matrix=cell_ids_matrix,
    cell_ids_harmonized=cell_ids_harmonized,
    dataset_id=ds["dataset_id"],
    source_accession=ds["source_accession"],
    chunksize=500,
)

adata_fixed.var_names = adata_fixed.var_names.astype(str).str.strip('"')
adata_fixed.var["gene_symbol"] = adata_fixed.var["gene_symbol"].astype(str).str.strip('"')
adata_fixed.var["gene_symbol_upper"] = adata_fixed.var["gene_symbol"].str.upper()
adata_fixed.var["gene_id"] = adata_fixed.var["gene_symbol"]
adata_fixed.var_names_make_unique()
adatas["GSE206330"] = adata_fixed

nb.summarize_adata(adata_fixed, name="GSE206330_fixed")
adata_fixed
```

```python
# ============================================================
# GSE206330 読み取り診断 + AnnData 再構築コード
# 前提:
#   import sys, Path, mf, man, paths, io_dense, au, nb は実行済み
#   ds = mf.get_dataset(man, "GSE206330")
# ============================================================

import gzip
import re
from pathlib import Path

import numpy as np
import pandas as pd
import scipy.sparse as sp
import anndata as ad


# ----------------------------
# 0. dataset / path確認
# ----------------------------
ds = mf.get_dataset(man, "GSE206330")

print("dataset keys:", list(ds.keys()) if isinstance(ds, dict) else type(ds))
print("dataset:", ds)

raw_dir = paths["raw"] / "GSE206330"
extracted_dir = paths["extracted"] / "GSE206330"

print("\nraw_dir:", raw_dir, "exists=", raw_dir.exists())
print("extracted_dir:", extracted_dir, "exists=", extracted_dir.exists())

print("\nraw files:")
for p in sorted(raw_dir.glob("*")):
    print(" ", p.name, p.stat().st_size / 1024 / 1024, "MB")

print("\nextracted files:")
if extracted_dir.exists():
    for p in sorted(extracted_dir.glob("**/*")):
        if p.is_file():
            print(" ", p.relative_to(extracted_dir), p.stat().st_size / 1024 / 1024, "MB")


# ----------------------------
# 1. GSE206330の対象ファイルを探す
# ----------------------------
def find_one_file(candidates, patterns):
    """
    candidates: list[Path]
    patterns: list[str]
    """
    hits = []
    for p in candidates:
        name = p.name
        for pat in patterns:
            if pat in name:
                hits.append(p)
                break
    hits = sorted(set(hits))
    if len(hits) == 0:
        raise FileNotFoundError(f"No file found for patterns={patterns}")
    if len(hits) > 1:
        print("Multiple hits:")
        for h in hits:
            print(" ", h)
        print("Using first:", hits[0])
    return hits[0]


all_files = []
if raw_dir.exists():
    all_files += [p for p in raw_dir.glob("*") if p.is_file()]
if extracted_dir.exists():
    all_files += [p for p in extracted_dir.glob("**/*") if p.is_file()]

count_path = find_one_file(
    all_files,
    patterns=["processed_SoupX_corrected.txt.gz", "processed_SoupX_corrected.txt"]
)

meta_path = find_one_file(
    all_files,
    patterns=["processed_SoupX_corrected_metadata.csv.gz", "processed_SoupX_corrected_metadata.csv"]
)

print("\ncount_path:", count_path)
print("meta_path :", meta_path)


# ----------------------------
# 2. 生ファイルの先頭を確認
# ----------------------------
def open_text_auto(path):
    path = Path(path)
    if path.suffix == ".gz":
        return gzip.open(path, "rt")
    return open(path, "rt")


print("\n=== raw count file first 5 lines, repr ===")
with open_text_auto(count_path) as f:
    for i in range(5):
        line = f.readline()
        print(f"[line {i+1}] len={len(line)}")
        print(repr(line[:1000]))
        print()

print("\n=== metadata first 5 lines, repr ===")
with open_text_auto(meta_path) as f:
    for i in range(5):
        line = f.readline()
        print(f"[line {i+1}] len={len(line)}")
        print(repr(line[:1000]))
        print()


# ----------------------------
# 3. 区切り文字を試す
# ----------------------------
def try_read_head(path, sep, header="infer", index_col=None, nrows=5, engine=None, label=""):
    print(f"\n--- try_read_head: {label} sep={repr(sep)} header={header} index_col={index_col} ---")
    try:
        kwargs = dict(sep=sep, header=header, index_col=index_col, nrows=nrows)
        if engine is not None:
            kwargs["engine"] = engine
        df = pd.read_csv(path, **kwargs)
        print("shape:", df.shape)
        print("columns[:10]:", list(df.columns[:10]))
        print("index[:5]:", list(df.index[:5]))
        display(df.iloc[:5, :min(10, df.shape[1])])
        return df
    except Exception as e:
        print("FAILED:", type(e).__name__, e)
        return None


# 典型パターンを全部試す
_ = try_read_head(count_path, sep="\t", header="infer", index_col=0, nrows=5, label="tab + header + index")
_ = try_read_head(count_path, sep=",",  header="infer", index_col=0, nrows=5, label="comma + header + index")
_ = try_read_head(count_path, sep=r"\s+", header="infer", index_col=0, nrows=5, engine="python", label="whitespace + header + index")

_ = try_read_head(count_path, sep=r"\s+", header=None, index_col=0, nrows=5, engine="python", label="whitespace + NO header + index")

meta_head = try_read_head(meta_path, sep=",", header="infer", index_col=0, nrows=5, label="metadata csv")


# ----------------------------
# 4. metadata全体を読む
# ----------------------------
meta = pd.read_csv(meta_path, index_col=0)
print("\n=== metadata full ===")
print("meta shape:", meta.shape)
print("meta index head:", list(meta.index[:5]))
print("meta columns:", list(meta.columns))
display(meta.head())


# ----------------------------
# 5. count matrixの形式を推定
# ----------------------------
# 1行目のtoken数、2行目のtoken数を見て header 有無を推定する
with open_text_auto(count_path) as f:
    line1 = f.readline().rstrip("\n")
    line2 = f.readline().rstrip("\n")

tokens1 = re.split(r"\s+", line1.strip())
tokens2 = re.split(r"\s+", line2.strip())

print("\n=== token diagnostics ===")
print("line1 token count:", len(tokens1))
print("line2 token count:", len(tokens2))
print("line1 first 10 tokens:", tokens1[:10])
print("line2 first 10 tokens:", tokens2[:10])
print("metadata rows:", meta.shape[0])

# 推定:
# - headerあり: 1行目が cell IDs header、2行目以降が gene + counts
# - headerなし: 各行が gene + counts で、列名はmetadataから補う
#
# よくある判定:
#   len(tokens2) - 1 == meta.shape[0] なら、gene + counts 形式
#   len(tokens1) - 1 == meta.shape[0] かつ tokens1[0]が空/遺伝子名っぽくないならheaderありの可能性
#
headerless_gene_by_cell = (len(tokens2) - 1 == meta.shape[0]) or (len(tokens1) - 1 == meta.shape[0])
print("headerless_gene_by_cell candidate:", headerless_gene_by_cell)

if len(tokens1) > 1:
    print("n numeric-like tokens in line1 after first token:",
          sum(pd.to_numeric(pd.Series(tokens1[1:50]), errors="coerce").notna()))


# ----------------------------
# 6. 小さく読み込んでmetadataと列数確認
# ----------------------------
# headerなし想定で読む
small_no_header = pd.read_csv(
    count_path,
    sep=r"\s+",
    engine="python",
    header=None,
    index_col=0,
    nrows=20
)

print("\n=== small_no_header ===")
print("shape:", small_no_header.shape)
print("index head:", list(small_no_header.index[:5]))
display(small_no_header.iloc[:5, :10])

print("small_no_header columns count:", small_no_header.shape[1])
print("metadata rows:", meta.shape[0])

# headerあり想定で読む
try:
    small_header = pd.read_csv(
        count_path,
        sep=r"\s+",
        engine="python",
        header=0,
        index_col=0,
        nrows=20
    )
    print("\n=== small_header ===")
    print("shape:", small_header.shape)
    print("index head:", list(small_header.index[:5]))
    print("columns head:", list(small_header.columns[:5]))
    display(small_header.iloc[:5, :10])
    print("intersection with metadata index:", len(set(small_header.columns) & set(meta.index)))
except Exception as e:
    small_header = None
    print("\nsmall_header FAILED:", type(e).__name__, e)


# ----------------------------
# 7. 読み込み方を決める
# ----------------------------
# 基本は以下：
# - headerなしで、count列数 == metadata行数なら header=None で読む
# - headerありで、columns と metadata.index が合うなら header=0 で読む
#
# この段階では巨大ファイルなので、いきなりdense DataFrame全体を作る前に
# chunkで sparse AnnData を作る関数を使う。

def build_adata_from_headerless_gene_by_cell_whitespace(
    count_path,
    meta,
    dataset_id="GSE206330_sc_cortex_sod1_facs_glia_soupx",
    source_accession="GSE206330",
    chunksize=1000,
):
    """
    count file:
      gene count count count ...
      gene count count count ...
    のような headerなし whitespace区切り matrix を想定。

    metadata rows = cells と仮定して、列順で対応づける。
    出力AnnDataは cells x genes。
    """
    X_blocks = []
    gene_names = []

    reader = pd.read_csv(
        count_path,
        sep=r"\s+",
        engine="python",
        header=None,
        index_col=0,
        chunksize=chunksize,
    )

    for i, chunk in enumerate(reader):
        # chunk: genes_chunk x cells
        if i == 0:
            print("first chunk shape:", chunk.shape)
            print("first chunk index head:", list(chunk.index[:5]))
            print("metadata rows:", meta.shape[0])

            if chunk.shape[1] != meta.shape[0]:
                raise ValueError(
                    f"Column count in matrix does not match metadata rows: "
                    f"chunk.shape[1]={chunk.shape[1]}, meta.shape[0]={meta.shape[0]}"
                )

        gene_names.extend(chunk.index.astype(str).tolist())

        # 数値化。失敗したらNaNになるので0埋め。
        chunk = chunk.apply(pd.to_numeric, errors="coerce").fillna(0)

        # AnnDataは cells x genes なので transpose
        X_blocks.append(sp.csr_matrix(chunk.T.values.astype(np.float32)))

        if (i + 1) % 10 == 0:
            print(f"  read chunks: {i+1}, genes so far: {len(gene_names)}")

    X = sp.hstack(X_blocks, format="csr")

    obs = meta.copy()
    obs.index = obs.index.astype(str)
    obs["cell_id_original"] = obs.index
    obs["sample_id"] = obs.get("Sample", obs.get("sample", "unknown"))
    obs["sample_label"] = obs["sample_id"].astype(str)
    obs["gsm_id"] = "unknown"
    obs["source_file"] = Path(count_path).name
    obs["source_accession"] = source_accession
    obs["dataset_id"] = dataset_id
    obs["data_status"] = "processed_SoupX_corrected"
    obs["processing_status"] = "SoupX_corrected"

    # meta列が上書きで衝突しないように、元metadata列をmeta_ prefixにも保持したい場合
    # ここでは既存列はそのまま残す。必要なら後段でrename。

    var = pd.DataFrame(index=pd.Index(gene_names, name=None).astype(str))
    var["gene_symbol"] = var.index
    var["gene_symbol_upper"] = var["gene_symbol"].str.upper()
    var["gene_id"] = var["gene_symbol"]
    var["ensembl_id"] = "unknown"
    var["feature_type"] = "unknown"

    adata = ad.AnnData(X=X, obs=obs, var=var)
    adata.obs_names = [
        f"{source_accession}_{x}" for x in adata.obs_names.astype(str)
    ]
    adata.obs["cell_id"] = adata.obs_names
    adata.var_names_make_unique()

    return adata


def build_adata_from_header_gene_by_cell_whitespace(
    count_path,
    meta,
    dataset_id="GSE206330_sc_cortex_sod1_facs_glia_soupx",
    source_accession="GSE206330",
    chunksize=1000,
):
    """
    count file:
      gene_or_blank cell1 cell2 cell3 ...
      gene count count count ...
    のような headerあり whitespace区切り matrix を想定。
    """
    # headerだけ読む
    header_df = pd.read_csv(
        count_path,
        sep=r"\s+",
        engine="python",
        header=0,
        index_col=0,
        nrows=1,
    )
    count_columns = header_df.columns.astype(str)

    common = pd.Index(count_columns).intersection(meta.index.astype(str))
    print("header columns:", len(count_columns))
    print("metadata rows:", meta.shape[0])
    print("intersection count:", len(common))

    X_blocks = []
    gene_names = []

    reader = pd.read_csv(
        count_path,
        sep=r"\s+",
        engine="python",
        header=0,
        index_col=0,
        chunksize=chunksize,
    )

    for i, chunk in enumerate(reader):
        if i == 0:
            print("first chunk shape:", chunk.shape)
            print("first chunk index head:", list(chunk.index[:5]))

        # metadataと共通するcellだけにする
        if len(common) > 0:
            chunk = chunk.loc[:, common]

        gene_names.extend(chunk.index.astype(str).tolist())
        chunk = chunk.apply(pd.to_numeric, errors="coerce").fillna(0)
        X_blocks.append(sp.csr_matrix(chunk.T.values.astype(np.float32)))

        if (i + 1) % 10 == 0:
            print(f"  read chunks: {i+1}, genes so far: {len(gene_names)}")

    X = sp.hstack(X_blocks, format="csr")

    if len(common) > 0:
        obs = meta.loc[common].copy()
    else:
        # headerありだがmetadataと合わない場合は列名をobsにする
        obs = pd.DataFrame(index=count_columns)

    obs.index = obs.index.astype(str)
    obs["cell_id_original"] = obs.index
    obs["sample_id"] = obs.get("Sample", obs.get("sample", "unknown"))
    obs["sample_label"] = obs["sample_id"].astype(str)
    obs["gsm_id"] = "unknown"
    obs["source_file"] = Path(count_path).name
    obs["source_accession"] = source_accession
    obs["dataset_id"] = dataset_id
    obs["data_status"] = "processed_SoupX_corrected"
    obs["processing_status"] = "SoupX_corrected"

    var = pd.DataFrame(index=pd.Index(gene_names, name=None).astype(str))
    var["gene_symbol"] = var.index
    var["gene_symbol_upper"] = var["gene_symbol"].str.upper()
    var["gene_id"] = var["gene_symbol"]
    var["ensembl_id"] = "unknown"
    var["feature_type"] = "unknown"

    adata = ad.AnnData(X=X, obs=obs, var=var)
    adata.obs_names = [
        f"{source_accession}_{x}" for x in adata.obs_names.astype(str)
    ]
    adata.obs["cell_id"] = adata.obs_names
    adata.var_names_make_unique()

    return adata


# ----------------------------
# 8. 実際にadata_fixedを作る
# ----------------------------
# 判定ロジック：
# small_no_header の列数が metadata rows と一致するなら headerなしとして読む。
# そうでなければ headerありとして読む。

if small_no_header.shape[1] == meta.shape[0]:
    print("\n>>> Using headerless whitespace gene x cell matrix")
    adata_fixed = build_adata_from_headerless_gene_by_cell_whitespace(
        count_path=count_path,
        meta=meta,
        dataset_id="GSE206330_sc_cortex_sod1_facs_glia_soupx",
        source_accession="GSE206330",
        chunksize=1000,
    )
else:
    print("\n>>> Using header whitespace gene x cell matrix")
    adata_fixed = build_adata_from_header_gene_by_cell_whitespace(
        count_path=count_path,
        meta=meta,
        dataset_id="GSE206330_sc_cortex_sod1_facs_glia_soupx",
        source_accession="GSE206330",
        chunksize=1000,
    )


# ----------------------------
# 9. 結果確認
# ----------------------------
adatas["GSE206330_fixed"] = adata_fixed

print("\n=== GSE206330 fixed ===")
nb.summarize_adata(adata_fixed, name="GSE206330_fixed")

display(adata_fixed)
display(adata_fixed.obs.head())
display(adata_fixed.var.head())

print("\nobs columns:")
print(list(adata_fixed.obs.columns))

print("\nvar columns:")
print(list(adata_fixed.var.columns))

print("\nX:")
print(type(adata_fixed.X), adata_fixed.X.shape, adata_fixed.X.dtype)
print("nnz:", adata_fixed.X.nnz if sp.issparse(adata_fixed.X) else np.count_nonzero(adata_fixed.X))
print("sum:", float(adata_fixed.X.sum()))

print("\nvalue counts for likely metadata columns:")
for col in ["Sample", "Condition", "Age", "cell_type", "sex",
            "meta_Sample", "meta_Condition", "meta_Age", "meta_cell_type", "meta_sex",
            "sample_id", "data_status", "processing_status"]:
    if col in adata_fixed.obs.columns:
        print(f"\n--- {col} ---")
        print(adata_fixed.obs[col].value_counts(dropna=False).head(30))


# ----------------------------
# 10. 必要なら一時保存
# ----------------------------
out_path = paths["interim"] / "GSE206330_sc_cortex_sod1_facs_glia_soupx.fixed_interim.h5ad"
print("\nwill save to:", out_path)

# 保存したくない場合は次の1行をコメントアウト
adata_fixed.write_h5ad(out_path)
print("saved:", out_path)
```
