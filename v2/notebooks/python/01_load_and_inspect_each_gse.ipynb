# 01 - 各GSEを AnnData 化して中身を確認

GSEごとに「ロード → その場で確認」を行う（旧 01+02 を統合）。生カウントは `adata.X`。ここでは機械的な obs（sample_id, source_file 等）だけ。genotype/treatment/condition は **02 で人間が設定**する。

先に download/extract を済ませること（`./run.sh` または `scripts/01..03`）。GSE295514 は先に R ノートブック `notebooks/R/01_GSE295514_read_rds.ipynb` を実行。

```python
import sys
from pathlib import Path

# config/dataset_manifest.yaml を持つ v2 ルートを探して src を import パスに追加
ROOT = Path.cwd()
while not (ROOT / "config" / "dataset_manifest.yaml").exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
print("project root:", ROOT)

import manifest_utils as mf
man = mf.load_manifest()
paths = mf.project_paths(ROOT)
mf.ensure_dirs(paths)

import io_10x
import io_dense
import anndata_utils as au
import notebook_report_utils as nb

adatas = {}
```

## GSE287569

10x .h5（TAR of H5）

```python
ds = mf.get_dataset(man, "GSE287569")
adata = io_10x.load_10x_h5_per_sample(ds, paths)
adatas["GSE287569"] = adata
nb.summarize_adata(adata, name="GSE287569")
adata
```

```python
# その場で確認
display(adata.obs.head())
display(adata.var.head())
_ = nb.show_obs_value_counts(adata, max_unique=30)
```
